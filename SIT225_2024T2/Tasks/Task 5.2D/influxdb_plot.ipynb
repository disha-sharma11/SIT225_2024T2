```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("/Users/dishasharma/Documents/Coding/SIT225/Task 5.2D/influxdata_2025-03-17T16_04_26Z.csv")
data["time"] = pd.to_datetime(data["time"])
```

```python
# Plot X-axis data
plt.figure(figsize=(8, 4))
plt.plot(data["time"], data["X"], label="Gyro X", color="r")
plt.xlabel("Time")
plt.ylabel("Angular Velocity")
plt.title("Gyroscope X-Axis Data")
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot Y-axis data
plt.figure(figsize=(8, 4))
plt.plot(data["time"], data["Y"], label="Gyro Y", color="g")
plt.xlabel("Time")
plt.ylabel("Angular Velocity")
plt.title("Gyroscope Y-Axis Data")
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot Z-axis data
plt.figure(figsize=(8, 4))
plt.plot(data["time"], data["Z"], label="Gyro Z", color="b")
plt.xlabel("Time")
plt.ylabel("Angular Velocity")
plt.title("Gyroscope Z-Axis Data")
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.plot(data["time"], data["X"], label="Gyro X", color="r")
plt.plot(data["time"], data["Y"], label="Gyro Y", color="g")
plt.plot(data["time"], data["Z"], label="Gyro Z", color="b")
plt.xlabel("Time")
plt.ylabel("Angular Velocity")
plt.title("Gyroscope Data (X, Y, Z)")
plt.legend()
plt.xticks(rotation=45)
plt.show()
```
