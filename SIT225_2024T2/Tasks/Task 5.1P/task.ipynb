```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("/Users/dishasharma/Documents/Coding/SIT225/Task 5.1P/gyroscope_data.csv")
data["timestamp"] = pd.to_datetime(data["timestamp"])
```

```python
# Plot X-axis data
plt.figure(figsize=(8, 4))
plt.plot(data["timestamp"], data["x"], label="Gyro X", color="r")
plt.xlabel("Time")
plt.ylabel("Angular Velocity")
plt.title("Gyroscope X-Axis Data")
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot Y-axis data
plt.figure(figsize=(8, 4))
plt.plot(data["timestamp"], data["y"], label="Gyro Y", color="g")
plt.xlabel("Time")
plt.ylabel("Angular Velocity")
plt.title("Gyroscope Y-Axis Data")
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot Z-axis data
plt.figure(figsize=(8, 4))
plt.plot(data["timestamp"], data["z"], label="Gyro Z", color="b")
plt.xlabel("Time")
plt.ylabel("Angular Velocity")
plt.title("Gyroscope Z-Axis Data")
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.plot(data["timestamp"], data["x"], label="Gyro X", color="r")
plt.plot(data["timestamp"], data["y"], label="Gyro Y", color="g")
plt.plot(data["timestamp"], data["z"], label="Gyro Z", color="b")
plt.xlabel("Time")
plt.ylabel("Angular Velocity")
plt.title("Gyroscope Data (X, Y, Z)")
plt.legend()
plt.xticks(rotation=45)
plt.show()
```
