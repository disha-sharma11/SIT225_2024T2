```python
import pandas as pd
import matplotlib.pyplot as plt

# Load all three datasets
x_data = pd.read_csv("/Users/dishasharma/Documents/Coding/SIT225/Task 3.1P/Task 3.1P-accelX.csv")
y_data = pd.read_csv("/Users/dishasharma/Documents/Coding/SIT225/Task 3.1P/Task 3.1P-accelY.csv")
z_data = pd.read_csv("/Users/dishasharma/Documents/Coding/SIT225/Task 3.1P/Task 3.1P-accelZ.csv")

# Convert time columns to datetime format
x_data["time"] = pd.to_datetime(x_data["time"])
y_data["time"] = pd.to_datetime(y_data["time"])
z_data["time"] = pd.to_datetime(z_data["time"])

# Merge all data into a single DataFrame on the 'time' column
combined_data = x_data.merge(y_data, on="time", suffixes=("_X", "_Y"))
combined_data = combined_data.merge(z_data, on="time")
combined_data.rename(columns={"value": "Accel_Z"}, inplace=True)

# Plot all accelerations in a single graph
plt.figure(figsize=(10, 5))
plt.plot(combined_data["time"], combined_data["value_X"], label="Accel X", color="r")
plt.plot(combined_data["time"], combined_data["value_Y"], label="Accel Y", color="g")
plt.plot(combined_data["time"], combined_data["Accel_Z"], label="Accel Z", color="b")

plt.xlabel("Time")
plt.ylabel("Acceleration")
plt.title("Combined Accelerometer Data Over Time")
plt.legend()
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```
